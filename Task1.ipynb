```python
#Task-1
def task(lst):
  ft=-1
  ans=[]
  while len(lst) > 0:
    x=heapq.heappop(lst)
    if x[1] >= ft:
      ans.append( ( x[1],x[0] ) )
      ft=x[0]
  return ans
```

```python
import heapq
inp=open('input1.txt')
outp=open('output1.txt','w')
line=inp.readline()
line=line.split('\n')
n=int(line[0])
lst=[]
for i in range(n):
  line=inp.readline()
  line=line.split('\n')
  line=line[0].split(' ')
  st=int(line[0])
  ed=int(line[1])
  lst.append( (ed,st) )
heapq.heapify(lst)
ans=task(lst)
print(len(ans))
outp.write(str(len(ans))+'\n')
for x in ans:
  print(x[0],x[1])
  outp.write(str(x[0])+' '+str(x[1])+'\n')
```
