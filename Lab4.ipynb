```python
#Task-1
import heapq
from collections import defaultdict
def Dijkstra(adj,n,st=1):
  pq=[(0,st)]
  cost=[100000 for i in range (n+1)]
  cost[st]=0
  heapq.heapify(pq)

  while len(pq)>0:
    x=heapq.heappop(pq)
    v=x[1]
    for y in adj[v]:
      u=y[0]
      c=y[1]
      if cost[v]+c < cost[u]:
        cost[u]=cost[v]+c
        heapq.heappush(pq,(cost[u],u) )
  return cost[n]
```

```python
inp=open('input1.txt')
outp=open('output1.txt','w')
tc=inp.readline()
tc=tc.split('\n')
tc=int(tc[0])
for i in range(tc):
  line=inp.readline()
  line=line.split('\n')
  line=line[0].split(' ')
  n=int(line[0])
  m=int(line[1])
  adj= defaultdict(list)
  for j in range(m):
    line=inp.readline()
    line=line.split('\n')
    line=line[0].split(' ')
    u=int(line[0])
    v=int(line[1])
    w=int(line[2])
    adj[u].append( (v,w) )
    adj[v].append( (u,w) )
  C=Dijkstra(adj,n)
  #print(C)
  outp.write(str(C)+"\n")
```

```python
#Task-2
def Dijkstra_Path(adj,n,st=1):
  par=[0]*(n+2)
  par[st]=st
  pq=[(0,st)]
  cost=[100000 for i in range(n+1)]
  cost[st]=0
  heapq.heapify(pq)

  while len(pq)>0:
    x=heapq.heappop(pq)
    v=x[1]
    for y in adj[v]:
      u=y[0]
      c=y[1]
      if cost[v]+c < cost[u]:
        cost[u]=cost[v]+c
        par[u]=v
        heapq.heappush(pq,(cost[u],u) )

  path=[n]
  curr=n
  while curr != par[curr]:
    curr=par[curr]
    path.append(curr)
  path.reverse()
  return path
```

```python
inp=open('input1.txt')
outp=open('output2.txt','w')
tc=inp.readline()
tc=tc.split('\n')
tc=int(tc[0])
for i in range(tc):
  line=inp.readline()
  line=line.split('\n')
  line=line[0].split(' ')
  n=int(line[0])
  m=int(line[1])
  adj= defaultdict(list)
  for j in range(m):
    line=inp.readline()
    line=line.split('\n')
    line=line[0].split(' ')
    u=int(line[0])
    v=int(line[1])
    w=int(line[2])
    adj[u].append( (v,w) )
    adj[v].append( (u,w) )
  path=Dijkstra_Path(adj,n)
  #print(path)
  for x in path:
    outp.write(str(x))
    if x==n:
      outp.write('\n')
    else :
      outp.write(' ')
```
